# Analyse Statistique

Utilisons des outils d'analyse statistique sur le jeu de données

```python
import pandas as pd
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
filename = '../Penn_State_PA_2012.gzip'
df = pd.read_csv(filename, encoding='utf-8', compression='gzip', index_col='timestamp', parse_dates=True)
df.index = df.index.tz_localize('UTC').tz_convert('US/Eastern')
```

```python
col_flag = df.columns[df.columns.str.endswith('_Flag')]
col_noflag = col_flag.str.replace('_Flag', '')
for i in col_noflag.values:
    df.loc[df[i + '_Flag'] > 0, i] = np.nan
```

```python
pivoted = df.pivot_table('diffuse',index=df.index.time,columns=df.index.date)
pivoted.head()
```

```python
sns.set_style("ticks")
sns.set_context("notebook")
ax = pivoted.plot(legend=False, alpha=0.006, ylim=[0,800], c='k')
sns.despine()
fig = ax.get_figure()
fig.savefig('../diffuse.png', transparent=True)
```

## Analyse de Composante Principale (PCA)
Une analyse de composante principale cherche a réduire un jeu de données à plusieurs dimensions (eg. 365 jours de 24h * 60 min = 1440 points)

```python
X = pivoted.fillna(0).T.values
X.shape
```

```python
from sklearn.decomposition import PCA
X2 = PCA(2).fit_transform(X)
```

```python
X2.shape
```

Nous obtenons une réduction de dimension pour 366 jours (2012 est une années bisecstiles) à deux dimensions, que nous pouvons visualiser sur un graph

```python
import matplotlib.pyplot as plt
plt.scatter(X2[:,0],X2[:,1])
plt.title('Principal Component Analysis')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(2,covariance_type='full')
gmm.fit(X)
labels = gmm.predict(X)
labels
```

```python
plt.scatter(X2[:,0],X2[:,1], c=labels, cmap='rainbow')
plt.colorbar()
```

```python
pivoted.T[labels==1].T.plot(legend=False, alpha=0.1)
```

```python
pivoted.T[labels==0].T.plot(legend=False, alpha=0.1)
```

```python
month = pd.DatetimeIndex(pivoted.columns).month
```

```python
plt.scatter(X2[:,0],X2[:,1], c=month, cmap='rainbow')
plt.colorbar()
```
