# Visualiser les données

Très rapidement, on peut visualiser les données en appelant la fonction `plot()`

```python
import pandas as pd
%matplotlib inline
```

```python
filename = '../Penn_State_PA_2012.gzip'
df = pd.read_csv(filename, encoding='utf-8', compression='gzip', index_col='timestamp', parse_dates=True)
df.index = df.index.tz_localize('UTC').tz_convert('US/Eastern')
```

```python
df.diffuse.plot()
```

On remarque que certaines données sont abérrantes (valeurs -9999.99). Sont le fichier de descriptions des données SURFRAD, elles correspondent à des données manquantes.

Voyons un peut plus en détail à quel moment de l'année elles interviennent.

```python
df['2012-06'].diffuse.plot()
```

Un intervalle de temps peut être sélectionné

```python
df['2012-06-23':'2012-06-30'].diffuse.plot()
```
